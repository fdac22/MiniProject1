```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

netflix_overall=pd.read_csv("netflix_titles.csv")

netflix_movies=netflix_overall[netflix_overall['type']=='Movie']
netflix_shows = netflix_overall[netflix_overall['type'] == 'TV Show']

plt.figure(figsize=(10,10))

wordcloud = WordCloud(stopwords=STOPWORDS, max_words=100).generate(' '.join(netflix_movies['title']))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Most Popular Words in Movie Titles',fontsize = 35)
plt.show()


plt.figure(figsize=(10,10))
wordcloud = WordCloud(stopwords=STOPWORDS, max_words=100).generate(' '.join(netflix_shows['title']))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Most Popular Words in TV Show Titles',fontsize = 35)
plt.show()
```

```python
from collections import Counter

genres=list(netflix_movies['listed_in'])
gen=[]

for i in genres:
    i=list(i.split(','))
    for j in i:
        gen.append(j.replace(' ',""))
g=Counter(gen)
g={k: v for k, v in sorted(g.items(), key=lambda item: item[1], reverse= True)}

plt.figure(figsize=(12,12))
plt.barh(list(g.keys()), list(g.values()))

plt.ylabel("Genres")
 
plt.xlabel("Number of Movies")
plt.title("Most Popular Movie Genres")
plt.show()

genres=list(netflix_shows['listed_in'])
gen=[]

for i in genres:
    i=list(i.split(','))
    for j in i:
        gen.append(j.replace(' ',""))
g=Counter(gen)
g={k: v for k, v in sorted(g.items(), key=lambda item: item[1], reverse= True)}

plt.figure(figsize=(12,12))
plt.barh(list(g.keys()), list(g.values()))

plt.ylabel("Genres")
 
plt.xlabel("Number of Shows")
plt.title("Most Popular TV Show Genres")
plt.show()
```
